```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import re
```

```python
train_data = pd.read_csv('train.csv')
print(train_data.shape)
print(train_data.describe())
selected_data = train_data.iloc[:101]

# Preprocess the training data
selected_data['ABSTRACT'] = selected_data['ABSTRACT'].str.lower()

selected_data['ABSTRACT'] = selected_data['ABSTRACT'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))

selected_data['ABSTRACT'] = selected_data['ABSTRACT'].apply(word_tokenize)

```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
selected_data['ABSTRACT'] = selected_data['ABSTRACT'].apply(lambda x: [word for word in x if word not in stop_words])
selected_data['ABSTRACT']
```

```python
stemmer = PorterStemmer()
selected_data['ABSTRACT'] = selected_data['ABSTRACT'].apply(lambda x: [stemmer.stem(word) for word in x])
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features = 500)
```

```python
# Transform the training data into TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(selected_data['ABSTRACT'].apply(lambda x: ' '.join(x)))
print("TF-IDF Matrix:")
print(tfidf_matrix.toarray())
tfidf_matrix.toarray()
```

```python
# Custom k-NN Classifier

import statistics

class KNNClassifiers:
    def __init__(self, k):
        self.k = k
    def euclidian_distance(self, x1, x2):
        # Assuming x1 and x2 are numpy arrays
        return np.sqrt(np.sum(np.square(x1 - x2)))

    
    def find_Nearest(self, x):
        try:
            euclidean_distances = np.zeros(self.m)
            for i in range(self.m):
                
                d = self.euclidian_distance(x, self.X_train.getrow(i).toarray()) 
                euclidean_distances[i] = d
            inds = euclidean_distances.argsort()
            y_train_sorted = self.y_train[inds]
            return y_train_sorted[:self.k]
        except Exception as e:
            raise e

    def fit(self, X_train, y_train):
        try:
            self.X_train = X_train 
            self.y_train = y_train
            self.m, self.n = X_train.shape
        except Exception as e:
            raise e
    
    def predict(self, X_test):
        try:
            self.X_test = X_test
            self.m_test, self.n_test = X_test.shape
            y_pred = np.zeros(self.m_test, dtype=object)
            for i in range(self.m_test):
                x = self.X_test.getrow(i).toarray()
                neighbours = self.find_Nearest(x)
                y_pred[i] = statistics.mode(neighbours) 
            return y_pred
        except Exception as e:
            raise e


class MyMultiOutputClassifier:
    def __init__(self, base_classifier_instance):
        self.base_classifier_instance = base_classifier_instance
        self.models = []
        
    def fit(self, X, y):
        self.models = []
        self.n_outputs = y.shape[1]
        
        for i in range(self.n_outputs):
            # Create a new instance of the base classifier for each output
            model = KNNClassifiers(self.base_classifier_instance.k)
            model.fit(X, y[:, i])
            self.models.append(model)
            
    def predict(self, X):
        predictions = np.zeros((X.shape[0], self.n_outputs))
        
        for i in range(self.n_outputs):
            predictions[:, i] = self.models[i].predict(X)
            
        return predictions

```

```python
import numpy as np
import matplotlib.pyplot as plt
def calculate_micro_f1_score(y_true, y_pred):
    true_positives = np.sum(np.logical_and(y_true, y_pred))
    false_positives = np.sum(np.logical_and(np.logical_not(y_true), y_pred))
    false_negatives = np.sum(np.logical_and(y_true, np.logical_not(y_pred)))
    
    precision = true_positives / (true_positives + false_positives)
    recall = true_positives / (true_positives + false_negatives)
    
    if precision + recall == 0:
        f1_score = 0.0
    else:
        f1_score = 2 * (precision * recall) / (precision + recall)
    
    return f1_score

def calculate_hamming_loss(y_true, y_pred):
    hamming_loss = np.mean(np.not_equal(y_true, y_pred))
    return hamming_loss
# Define a range of K values for the KNN algorithm
knn_k_values = [5, 10, 15, 20, 25, 30]  # Adjust as needed

# Initialize lists to store average Hamming Loss and F1 scores for each K value
average_hamming_losses = []
average_f1_scores = []

# Number of splits (k) for cross-validation (remains constant)
k_fold = 5

# Custom k-fold cross-validation
n_samples = tfidf_matrix.shape[0]
fold_size = n_samples // k_fold

for knn_k in knn_k_values:
    # Initialize lists to store Hamming Loss and F1 scores for each fold
    hamming_losses = []
    f1_scores = []
    for i in range(k_fold):
        # Calculate fold indices
        start = i * fold_size
        end = (i + 1) * fold_size if i < k_fold - 1 else n_samples

        # Split data into train and test sets
        test_indices = range(start, end)
        train_indices = [j for j in range(n_samples) if j not in test_indices]

        X_train_fold, X_test_fold = tfidf_matrix[train_indices], tfidf_matrix[test_indices]
        y_train_fold, y_test_fold = y_train_np[train_indices], y_train_np[test_indices]

        # Instantiate your custom multi-output classifier
        multi_output_classifier = MyMultiOutputClassifier(KNNClassifiers(knn_k))

        # Fit the model on the training fold
        multi_output_classifier.fit(X_train_fold, y_train_fold)

        # Make predictions on the test fold
        y_pred_fold = multi_output_classifier.predict(X_test_fold)
        print(f"y_pred for {i+1}th fold: ")
        print(y_pred_fold)

        # Calculate Hamming Loss and F1 score for the fold
        hamming_loss_fold = calculate_hamming_loss(y_test_fold, y_pred_fold)
        f1_score_fold = calculate_micro_f1_score(y_test_fold, y_pred_fold)

        # Store Hamming Loss and F1 score for this fold
        hamming_losses.append(hamming_loss_fold)
        f1_scores.append(f1_score_fold)

    # Calculate and store the average Hamming Loss and F1 Score for this K value
    average_hamming_loss = sum(hamming_losses) / len(hamming_losses)
    average_f1_score = sum(f1_scores) / len(f1_scores)

    average_hamming_losses.append(average_hamming_loss)
    average_f1_scores.append(average_f1_score)

    print(f'Average Hamming Loss (k = {knn_k}): {average_hamming_loss:.2f}')
    print(f'Average F1 Score for k={knn_k} is: {average_f1_score:.2f}')

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(knn_k_values, average_hamming_losses, marker='o', label='Average Hamming Loss')
plt.plot(knn_k_values, average_f1_scores, marker='o', label='Average F1 Score')
plt.xlabel('KNN K (Number of Neighbors)')
plt.ylabel('Metric Value')
plt.title('KNN Performance for Different K Values')
plt.legend()
plt.grid(True)
plt.show()
```
